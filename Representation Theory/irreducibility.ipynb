# Test for irreducability in a representation

- Take a representation (group elements mapped to matricies -> array of matricies built from generators)
    - generators -> array of matricies p(g) w/ dimension n

- Find the commutant of the representation (the set of matricies that commutes w/ the representation -> another array of matricies)
    - build commuting condition
    - for every generator g, calculate $M_g = I \otimes p(g) - p(g)^T \otimes I$
        - This encodes the idea of commutation into M_g for every generator g
    - Stack all of the M_g's into a tall matrix M

- Find the dimension of the commutant, and it will tell you if the representation is irreducable (dim Com = 1) or not -> compute nullity of the vectorized commutant
    - dim(Commutant) = nullity(M) = $n^2$ - rank(M)
        - dim(Comm) = 1 -> irreducable
        - dim(Comm) > 1 -> reducable

Consider the cyclic group $\mathbb{Z}_n$, the integers modulo n

This group only has a single generator, with every element being a power of that generator, shifting the basis by 1 cyclically. The representation is 
$$ \rho : G \to GL(V) $$
where $$V = \mathbb{C}^n$$

The single generator of this group is the element $1 \in \mathbb{Z}_n$, the abstract idea of 'add 1 mod n'. This element produces the rest of the elements

$\rho(1)$ is the matrix that the representation assigns to the generator, the cyclic shift generator shown below

```python
# representation matrix of the single generator that generates every element of Zn
p1 = np.array([
    [0, 0, 1],
    [1, 0, 0],
    [0, 1, 0]
], astype=complex)

# for S3
s1 = np.array([
    [0, 1, 0],
    [1, 0, 0],
    [0, 0, 1]
], astype=complex)
s2 = p1


# Z4, S4
```

Since every other element is generated by the same matrix 

(powers of $\rho, \rho(2) = \rho(1)^2, \rho(3) = \rho(1)^3$, etc.)

If a matrix commutes with $\rho(1)$, it commutes with every power.

```python
generators = [p1]
n = p1.shape[0]
I = np.eye(n, astype=complex)

# then calculate the vectorized commutant

M = np.vstack([np.kron(I, g) - np.kron(np.transpose(g), I) for g in generators], astype=complex)

M
```

dim Commutant = nullity(M) = $n^2$ - rank(M)

To get the dimension of the commutant, find the nullity of M, or the dimension of the nullspace/kernel of constraint matrix. -> $n^2$ is the dimension of the whole space of matricies (commuting or not), so to get the ones that commute, is the kernel of the matrix that tests the commuting property for all of the generators, as the one's that map to 0 commute. 

```python
# now compute the nullity of M to find the dimension of the commutant
dimC = n**2 - np.linalg.matrix_rank(M)

print(f"The dimension of the commutant is {dimC}")
print(f"The representation is irreducible" * int(dimC==1))
print(f"The representation is reducible" * int(dimC!=1))
```
